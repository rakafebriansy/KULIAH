# TEKNIK PLOTTING

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#VERTICAL BAR GRAPHS#
values = [200,320,440]
labels = ["Faqih","Raka","Arie"]

plt.bar(labels,values,color="blue")
plt.xlabel("Jumlah Suara")
plt.xlabel("Nama Calon Ketua OSIS")
plt.title("Pemilihan Kandidat Calon Ketua BEM")
plt.show()
```

```python
#HORIZONTAL BAR GRAPHS#
values = [40963227,36896752,46092205,15032170,11100929]
labels = ["Jawa Timur","Jawa Tengah","Jawa Barat","Sumatera Utara","DKI Jakarta"]

plt.barh(labels,values,color="green")
plt.xlabel("Jumlah Penduduk")
plt.xlabel("Nama Provinsi")
plt.title("Provinsi Dengan Penduduk Terbanyak")
plt.show()
```

```python
#PIE CHART#
labels = ["Bermain","Belajar","Makan","Tidur"]
slices = [6,9,2,7]
colors = ["r","g","b","y"]
plt.pie(slices,labels=labels, colors=colors,startangle=90,shadow=True,explode=(0,0.1,0,0),radius=1.2,autopct="%1.1f%%")
plt.show()
```
