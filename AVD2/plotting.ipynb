# TEKNIK PLOTTING

```python
import matplotlib.pyplot as plt, seaborn as sns, pandas as pd
```

```python
pelanggan = pd.read_csv("pelanggan.csv")
negara = pd.read_csv("negara.csv")
kartu_kredit = pd.read_csv("kartu_kredit.csv")
food_sales = pd.read_csv("food_sales.csv",delimiter=";")
```

```python
#VERTICAL BAR GRAPHS#

values = [negara.iloc[70,5],negara.iloc[95,5],negara.iloc[133,5]]
labels = [negara.iloc[70,0],negara.iloc[95,0],negara.iloc[133,0]]

plt.bar(labels,values,color="blue")
plt.xlabel("Negara")
plt.ylabel("Jumlah Pendapatan")
plt.title("Perbandingan Jumlah Pendapatan Antara Indonesia dengan Dua Negara Tetangganya")
plt.show()
```

```python
#HORIZONTAL BAR GRAPHS#
values = [negara.iloc[70,9],negara.iloc[95,9],negara.iloc[133,9],negara.iloc[148,9],negara.iloc[122,9],negara.iloc[107,9],negara.iloc[27,9],negara.iloc[84,9],negara.iloc[164,9],negara.iloc[23,9],negara.iloc[149,9]]
labels = [negara.iloc[70,0],negara.iloc[95,0],negara.iloc[133,0],negara.iloc[148,0],negara.iloc[122,0],negara.iloc[107,0],negara.iloc[27,0],negara.iloc[84,0],negara.iloc[164,0],negara.iloc[23,0],negara.iloc[149,0]]

plt.barh(labels,values,color="green")
plt.xlabel("Negara")
plt.ylabel("Gross Domestic Product (GDP)")
plt.title("Perbandingan GDP Negara-negara di Asia Tenggara")
plt.show()
```

```python
#PIE CHART#
labels = ["Ekspor","Impor"]
slices = [negara.iloc[70,2],negara.iloc[70,4]]
colors = ["r","g"]
plt.pie(slices,labels=labels, colors=colors,startangle=90,shadow=True,radius=1.2,autopct="%1.1f%%")
plt.title("Persentase Ekspor dan Impor Negara Indonesia")
plt.show()
```

```python
#BOXPLOT#
sns.boxplot(data=pelanggan, x='Annual Income (k$)', y='Gender')
plt.title("Sebaran Data Pendapatan Berdasarkan Jenis Kelamin")
plt.show()
```

```python
#HISTOGRAM#
cust_age = pelanggan["Age"]
plt.hist(cust_age)
plt.title("Sebaran Data Umur Pelanggan")
plt.show()
```

```python
#KERNEL DENSITY ESTIMATION (KDE)#
cust_age = pelanggan["Age"]
ax = cust_age.plot.kde()
plt.title("Estimasi Kepadatan Data Umur Pelanggan")
plt.xlabel("Umur Pelanggan")
plt.show()
```
